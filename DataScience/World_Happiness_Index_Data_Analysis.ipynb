## Abstract

EDA 프로젝트의 한 예로써 World Happniness Report에 대하여 실습을 하고자 한다. 이 실습에서 사용하는 데이터는 아래와 같다.

World Happiness Report는 글로벌 행복지수를 여섯개의 지표 - 건강 기대 수명, 1인당 국내총생산(경제력), 사회적 지원, 부정부패 인식, 삶의 선택자유, 공동체 나눔(관용)를 추적한다.

이 데이터 세트를 분석함으로써 다음의 질문에 대하여 답을 할 수 있다:

* 전 세계에서 가장 행복한 나라는 어디인가?
* 국민의 행복을 위해서 가장 중요한 지표는 무엇인가?
* 전반적인 행복지수는 증가하고 있는가? 감소하고 있는가?

본 실습을 통해 Kaggle로 부터 공개 데이터를 획득하는 방법을 배우며, 분석가로써 필요한 스킬을 향상시킬 수 있다. 가장 성공적인 데이터 분석가가 가져야 하는 것은 호기심이다. 

데이터 과학자와 데이터 분석가는 항상 기여 요인(contributing factors)을 찾고 있어야 한다. 항상 변수 간의 관계를 찾고 있으며 끊임없이 질문을 한다. 만약 여러분이 장차 데이터 과학자가 되고 싶다면 이번 프로젝트가 분석가의 마인드를 개발하는 데 도움이 될 것이다.

* **Parallel coordinates** : This plot shows the relationship between the adjacent axis(only adjacent axis). One disadvantage here is each axis has a different scale, so it’s better we adjust the scale so that relationship between the adjacent axis is represented clearly.

```python
def zscore(series):
    return ((series - series.mean(skipna=True)) / series.std(skipna=True))

dataEU = data[data['Region'].isin(EU)].copy()
for col in cols:
    dataEU[col] = zscore(dataEU[col]) 

plt.figure(figsize=(10,5))
pc = parallel_coordinates(dataEU[cols + ['Region']], 'Region', color=EUcol)
```

**[프로젝트2 확장]**

2015년도 행복지수 1등의 국가는 각각의 수치별로 모든 국가의 평균보다 몇 배 더의 수치를 지니고 있는지 궁금하였습니다.

(모든 국가들의 평균보다 얼마나 더 행복한지 궁금하였습니다.)

cf) 해당 프로그램은 14번째 코드까지 실행시킨 후 실행하여야 오류가 발생하지 않습니다.

```python
data = pd.read_csv('2015.csv')

#2015년도 각 수치별 평균
data_mean = []
#유의미한 값만 도출하기 위해 cnt값 이용하여 인덱스 조작
cnt = 0
for p in data.mean():
  cnt += 1
  if cnt < 4:
    continue
  data_mean.append(p)
#각 col의 이름
attrName = []
cnt = 0
for i in data.loc[data['Happiness Rank'] == 1]:
  cnt += 1
  if cnt < 6:
    continue
  attrName.append(i)
#행복지수 1등 국가의 수치들
winnerData = []
for i in attrName:
  for j in data[i]:
    winnerData.append(j)
    break
#행복지수 1등 국가가 다른 국가들보다 얼마나 행복한지(비율)
result = []
for i in range(7):
  result.append(winnerData[i]/data_mean[i])

#print(attrName)
#print(result)

ax = sns.barplot(attrName, result)
ax.set_title("How happier the happiest country is than the average of other countries in 2015")
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
```
